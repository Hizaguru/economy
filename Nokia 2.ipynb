```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from yahooquery import Ticker
import csv
plt.style.use("seaborn-v0_8")
```

```python
symbol = "NOK" 
```

```python
nok = yf.Ticker(ticker = symbol)
nok
nok.history()
```

```python
nok_info = nok.get_info()
#current_ratio = nok_info['currentRatio']
#print("Current ratio", current_ratio)

# Get the dictionary from nok.get_info()
nok_info = nok.get_info()

# Create a pandas dataframe from the dictionary
df = pd.DataFrame.from_dict(nok_info)

# Export the dataframe to an Excel file
df.to_csv('nok_info.csv', index=False)
```

```python
with open('dict.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in yq.key_stats['NOK'].items():
       writer.writerow([key, value])
```
