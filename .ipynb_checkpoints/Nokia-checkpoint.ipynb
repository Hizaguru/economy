```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
symbol = "NOK" 
```

```python
df  = yf.download(tickers = symbol)
df
```

```python
df.index
```

```python
df.info()
```

```python
# Price chart

df.Close.plot(figsize = (12,8), fontsize =12)
plt.ylabel("Price (USD)")
plt.title("Nokia price chart", fontsize = 15)
plt.show()
```

```python
#Volume chart
df.loc("2023-4", "Volume").plot(figsize = (12,8) kind= "bar")
plt.ylabel("Volume (shares)")
plt.title("Nokia trading volume (daily)", fontsize=15)
plt.show()
```
